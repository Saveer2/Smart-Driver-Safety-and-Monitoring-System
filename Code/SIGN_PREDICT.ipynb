```python
import sys
sys.stdout.reconfigure(encoding='utf-8')

import os
os.environ["PYTHONIOENCODING"] = "utf-8"

import cv2
import numpy as np
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Load model
model = load_model("traffic_sign_model.h5")

# Class labels
classes = {
0:'Speed Limit 20', 1:'Speed Limit 30', 2:'Speed Limit 50',
3:'Speed Limit 60', 4:'Speed Limit 70', 5:'Speed Limit 80',
6:'End of Speed Limit 80', 7:'Speed Limit 100', 8:'Speed Limit 120',
9:'No passing', 10:'No passing vehicles > 3.5 tons',
11:'Right-of-way at intersection', 12:'Priority road',
13:'Yield', 14:'Stop', 15:'No vehicles',
16:'Vehicles > 3.5 tons prohibited', 17:'No entry',
18:'General caution', 19:'Dangerous curve left',
20:'Dangerous curve right', 21:'Double curve',
22:'Bumpy road', 23:'Slippery road', 24:'Road narrows',
25:'Road work', 26:'Traffic signals', 27:'Pedestrians',
28:'Children crossing', 29:'Bicycles crossing',
30:'Beware of ice/snow', 31:'Wild animals crossing',
32:'End speed + passing limits', 33:'Turn right ahead',
34:'Turn left ahead', 35:'Ahead only',
36:'Go straight or right', 37:'Go straight or left',
38:'Keep right', 39:'Keep left',
40:'Roundabout mandatory', 41:'End of no passing',
42:'End no passing vehicles > 3.5 tons'
}

# Image path
image_path = "test.jpg"

# Read image
image = cv2.imread(image_path)

if image is None:
    print("Image not found")
else:
    img = cv2.resize(image, (30,30))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)

    predictions = model.predict(img, verbose=0)
    class_id = np.argmax(predictions)
    confidence = np.max(predictions)

    label = classes[class_id]

    print("Prediction:", label)
    print("Confidence:", round(confidence, 2))

    # Show image using matplotlib (fix for cv2 error)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"{label} ({confidence:.2f})")
    plt.axis('off')
    plt.show()
```
